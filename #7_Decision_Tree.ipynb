# Disadvantage :-

> It learn the data completlly that it not good perform on testing.That why it get over fit most of the time.

> It good for logistic regression but not for linear regression.

> QUE:- How we can mange the overfitting in the Decision tree?

> Ans> by using leveling  and stop the algo. at mid level. 

```python
import numpy as np
import pandas as pb
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=pb.read_csv(r"A:\Ducat\PYTHON\phyton\data analytics\MACHINE LEARNING\DATASETS\IRIS.csv")
```

```python
df.species.value_counts()
```

```python
feature=df.iloc[:,:4].values
print("dimension :- ",feature.ndim,"\n",feature) #2d array
```

```python
target=df.species
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(feature,target)
```

```python
plt.figure(figsize=(15,10))
plot_tree(model)
plt.show()
```

```python
print(model.predict([[5.4 ,3.9 ,1.7 ,0.4]]),
model.predict_proba([[5.4 ,3.9 ,1.7 ,0.4]])) ## this condition is in overfitting.
```

# -----> TOO solve this we use leveling.:-

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
model1=DecisionTreeClassifier(max_depth=3)
```

```python
model1.fit(feature,target)
```

```python
plt.figure(figsize=(15,10))
plot_tree(model1)
plt.show()
```

```python
print(model1.predict([[5.4 ,3.9 ,1.7 ,0.4]]),"\n",model1.predict_proba([[5.4, 3.9, 1.7, 0.4]])) 
```
