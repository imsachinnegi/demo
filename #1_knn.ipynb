 ## NOTE :- we divide the data into parts lable (1-d array)  and features (2-d array)  *Must

```python
y=df['Class']
y
```

```python
x=df.drop(['Class'],axis=1)
x
```

```python
from sklearn.neighbors import KNeighborsClassifier 
model=KNeighborsClassifier()
model
```

```python
model.fit(x,y)
```

```python
model.predict([[655,4.5555]])
```
